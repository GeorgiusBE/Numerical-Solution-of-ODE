In this project, I will be solving a second-order linear ordinary differential equation using numerical method (with the help of Taylor series expansion).

The following general form will be used:

$$\frac{d^2 y}{dx^2} + P(x) \frac{dy}{dx} + Q(x) y = f(x) \tag{1}$$

With the boundary conditions $y(a) = \alpha$ and $y(b) = \beta$

Let's denote

$$\delta x = \frac{b - a}{n}$$

Where n is the number of partitions between $a$ and $b$.

Moreover, for convinience, let

$$ y_i = y(x_i) , P_i = P(x_i) , Q_1 = Q(x_i) , f_i = f(x_i) $$

The numerical teachnique involves the following steps:
1. Select the appropriate number of steps (n).
2. Numerically approximate the derivatives in the ODE using Taylor series expansion (central difference).
3. Convert the equations that you obtained from step 2 into a matrix inversion problem that takes the form Ax = b.
4. Solve for the vector x.

#### Defining a function to numerically solve the ODE

Note that in the function below, LU (Lower-Upper) decomposition is used to solve the matrix equation above.

```python
def numerical_ode(a,b,n,lower_bound,upper_bound):
    '''
    a -> the lower bound for value of x
    b -> the upper bound for value of x
    n -> number of steps
    lower_bound -> the lower bound for value of y
    upper-boundary -> the upper bound for value of y
    '''
    # step size
    dx = (b-a)/n

    # range of values of x (excluding x at the boundaries)
    x = np.zeros(n-1)
    for step in range(1,n):
        x[step-1] = a + step*dx

    # sup-diagonal of matrix A
    P = 3
    sup_A = 1 + dx/2 * P

    # diagonal of matrix A
    Q = 2
    diag_A = -2 + Q*dx**2

    # sub-diagonal of matrix A
    sub_A = 1 - dx/2 * P

    # vector b
    f_vec = (4 * x**2)

    boundary_vec = np.zeros(n-1)
    boundary_vec[0] = (1 - dx/2 * P) * lower_bound
    boundary_vec[-1] = (1 + dx/2 * P) * upper_bound

    b_vec = f_vec* dx**2 - boundary_vec

    # Perform LU decomposition of matrix A
    u = np.zeros(n-2)
    d = np.zeros(n-1)
    l = np.zeros(n-2)

    d[0] = diag_A
    for i in range(n-2):
        u[i] = sup_A
        l[i] = sub_A/d[i]
        d[i+1] = diag_A - l[i] * u[i]
        
    # Solve the matrix equation Ax = b
        # Solve Lw = b
    w_vec = np.zeros(n-1)
    w_vec[0] = b_vec[0]/1
    for i in range(1,n-1):
        w_vec[i] = b_vec[i] - l[i-1] * w_vec[i-1]

        # Solve Ux = w
    solution = np.zeros(n-1)
    solution[-1] = w_vec[-1]/d[-1]
    for i in np.arange(-2,-(n-1)-1,-1):
        solution[i] = (w_vec[i] - u[i+1] * solution[i+1])/d[i]
        
    # range of values of x (including x at the boundaries)
    complete_xrange = np.append(x, 2)
    complete_xrange = np.insert(complete_xrange, 0, 1)

    # range of values of the solution y (including y at the boundaries)
    complete_yrange = np.append(solution, 6)
    complete_yrange = np.insert(complete_yrange, 0, 1)

    # Present the solution
    df_solution = pd.DataFrame({"x_i":complete_xrange, "y_i":complete_yrange})
    
    return df_solution
```

#### The exact solution

```python
def analytical_ode(a,b):
    n = 200
    dx = (b-a)/200
    
    df_solution = pd.DataFrame({"x_i":np.arange(a,b+dx,dx)})
    index = 0
    for i in np.arange(a,b+dx,dx):
        y = 38.222*np.e**(-i) - 118.6765*np.e**(-2*i) + 2*i**2 - 6*i + 7
        df_solution.loc[index,"y_i"] = y
        index = index + 1
        
    return df_solution
```

This means that the accuracy of the numerical solution increases as we increase the value of n.

#### Plotting the solution y against the number of steps (n)

```python
# This function calculates the numerical solution when x=1.5 with different values of n
def vary_n(a,b,lower_bound,upper_bound,lower_n,upper_n):
    '''
    lower_n -> lower boundary for n
    upper_n -> upper boundary for n
    '''
    # Range of values for n
    n_range = np.arange(lower_n,upper_n+2,2)
    
    # Dataframe to store solution
    df = pd.DataFrame({"n":n_range, "Numerical Solution":np.zeros(len(n_range))})
    
    idx = 0
    for n in n_range:
        num_ode = numerical_ode(a,b,n,lower_bound,upper_bound)
        select = num_ode["x_i"]==1.50
        df.loc[idx,"Numerical Solution"] = float(num_ode.loc[select,"y_i"])
        idx = idx + 1
    
    return df
```

```python
# The exact solution when x=1.5
exact = analytical_ode(a=1,b=2)
float(exact.loc[round(exact["x_i"],5)==1.5,"y_i"])
```

```python
# Calculates the numerical solution when x=1.5 with different values of n
df_n = vary_n(a=1,b=2,lower_bound=1,upper_bound=6,lower_n=10,upper_n=100)

# the exact solution when x=1.5
exact = analytical_ode(a=1,b=2)
exact = float(exact.loc[round(exact["x_i"],5)==1.5,"y_i"])

# Plot
fig, ax = plt.subplots(figsize=(7,5))

ax.plot(df_n["n"], df_n["Numerical Solution"], label="Numerical Solution")
ax.axhline(exact, color="#fc2412",ls='--',linewidth = '1', label="Exact Solution")

ax.set_xlabel("Number of Steps (n)")
ax.set_ylabel("y")
ax.set_title("Numerical Solution when x=1.5 with Varying n")
ax.legend()

plt.plot()
```

From the plot above, we notice that as the number of steps increases (n), the numerical solution line gets closer to the exact solution. This means that we can increase the accuracy of our numerical method by increasing n (which is consistent with what I have mentioned before).
